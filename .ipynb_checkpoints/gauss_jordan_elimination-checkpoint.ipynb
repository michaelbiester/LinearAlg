# Gauss-Jordan Elimination

In this notebook I follow the steps described in the book

`No Bullshit guide to Linear Algebra` , Ivan Savov



The three equations

$$\begin{align}
1x +  2y + 3z &= 14 \\
2x +  5y + 6z &= 30 \\
-1x +  2y + 3z &= 12
\end{align}
$$

which is cast to the augmented matrix equation:

$$
\left[
\begin{array}{ccccc}
1 & 2 & 3 & | & 14 \\
2 & 5 & 6 & | & 30 \\
-1 & 2 & 3 & | & 12 
\end{array}
\right]
$$

The next steps are to obtain the equation in `REF` (row-echelon form).

**Step#1:** row operation $R_2 - 2R_1 \to R_2$

$$
\left[
\begin{array}{ccccc}
1 & 2 & 3 & | & 14 \\
0 & 1 & 0 & | & 2 \\
-1 & 2 & 3 & | & 12 
\end{array}
\right]
$$

**Step#2:** row operation $R_3 + R_1 \to R_3$

$$
\left[
\begin{array}{ccccc}
1 & 2 & 3 & | & 14 \\
0 & 1 & 0 & | & 2 \\
0 & 4 & 6 & | & 26 
\end{array}
\right]
$$

**Step#3:** row operation $R_3 - 4R_2 \to R_3$

$$
\left[
\begin{array}{ccccc}
1 & 2 & 3 & | & 14 \\
0 & 1 & 0 & | & 2 \\
0 & 0 & 6 & | & 18 
\end{array}
\right]
$$

**Step#4:** $(1/6) R_3 \to R_3$

$$
\left[
\begin{array}{ccccc}
1 & 2 & 3 & | & 14 \\
0 & 1 & 0 & | & 2 \\
0 & 0 & 1 & | & 3 
\end{array}
\right]
$$

At this point the equation system has been transformed into `REF` form.

In the next steps the equation system is transformed into `RREF` (reduced row echelon form).

Starting with the 3'rd row we perform these steps:

**Step#1:** $R_1 - 3R_3 \to R_1$

$$
\left[
\begin{array}{ccccc}
1 & 2 & 0 & | & 5 \\
0 & 1 & 0 & | & 2 \\
0 & 0 & 1 & | & 3 
\end{array}
\right]
$$

**Step#2:** $R_1 - 2R_2 \to R_1$

$$
\left[
\begin{array}{ccccc}
1 & 0 & 0 & | & 1 \\
0 & 1 & 0 & | & 2 \\
0 & 0 & 1 & | & 3 
\end{array}
\right]
$$

At this point the equation system has been transformed into `RREF` form from which the solution can be directly obtained:

$x_1 = 1,\ x_2 = 2,\ x_3 = 3$

----


## Number of solutions

For a equation systems that can be represented by a $3 \times 3$ matrix there several cases to consider:

1) there is one solution

2) there infinitely many solutions in 1 dimensions

3) there infinitely many solution in 2 dimensions

4) there are no solutions

**One Solution**

The augmented matrix shall be `RREF` form.

$$
\left[
\begin{array}{ccccc}
1 & 0 & 0 & | & c_1 \\
0 & 1 & 0 & | & c_2 \\
0 & 0 & 1 & | & c_3 
\end{array}
\right]
$$

with solution $x_1 = c_1,\ x_2 =\ c_2,\ x_3 =\ c_3$.

**Infinitely many solutions / 1 dimensions**

We assume that the transformation to `REF` yields the 3'rd row with all zeros.

$$
\left[
\begin{array}{ccccc}
1 & 0 & a_1 & | & c_1 \\
0 & 1 & a_2 & | & c_2 \\
0 & 0 & 0 & | & 0 
\end{array}
\right]
$$

We are then free to choose $x_3 = t$ with $t$ being an arbitrary number. 

The system of equations is now expressed like this:

$$\begin{align}
x_1 +  a_1 \cdot t &= 14 \\
2 \cdot x_1 +  5 \cdot x_2 + 6 \cdot t &= 30 \\
x_3 &= t
\end{align}
$$

or in vector notation:

$$
\left[\begin{array}{c}
x_1 \\ x_2 \\ x_3
\end{array}\right] = \left[\begin{array}{c}
c_1 \\ c_2 \\ 0
\end{array}\right] + t \cdot \left[\begin{array}{c}
-a_1 \\ -a_2 \\ 1
\end{array}\right]
$$

This the vectorial representation of a line passing through point $[c_1,\ c_2,\ 0]$ in the direction of vector $[-a_1,\ -a_2,\ 1]$.


**No solutions**

$$
\left[
\begin{array}{ccccc}
1 & 0 & 0 & | & c_1 \\
0 & 1 & 0 & | & c_2 \\
0 & 0 & 0 & | & c_3 
\end{array}
\right]
$$

While $x_1 = c_1,\ x_2 = c_2$ this choice of $x_1,\ x_2$ does not satisfy the equation from the 3'rd row

$$
x_1 \cdot 0 + x_2 \cdot 0 + x_3 \cdot 0 = c_3
$$

(assuming $c_3 \neq 0$)


## Solving Excercises

**E.3.1**

augmented matrix:

$$
\left[
\begin{array}{ccc}
3 & 3 & | & 6 \\
2 & 1.5 & | & 5
\end{array}
\right]
$$

Step: $(1/3)*R_1 \to R_1$

$$
\left[
\begin{array}{ccc}
1 & 1 & | & 2 \\
2 & 1.5 & | & 5
\end{array}
\right]
$$

Step: $R_2 - 2*R_1 \to R_2$

$$
\left[
\begin{array}{ccc}
1 & 1 & | & 2 \\
0 & -0.5 & | & 1
\end{array}
\right]
$$

Step: $-2 * R_2 \to R_2$

$$
\left[
\begin{array}{ccc}
1 & 1 & | & 2 \\
0 & 1 & | & -2
\end{array}
\right]
$$


Step: $R_1 - R_2 \to R_1$

$$
\left[
\begin{array}{ccc}
1 & 0 & | & 4 \\
0 & 1 & | & -2
\end{array}
\right]
$$

Solution: $x_1 = 4,\ x_2 = -2$

# Solution to excersises 

Excersises are from the book

`No Bullshit guide to Linear Algebra` , Ivan Savov; Chapter 3.6 (Computational problems)

**P.3.4**
Find the solution sets for the augemented matrices.

a)

$$
\left[
\begin{array}{cccc}
-1 & -2 & | & -2 \\
3 & 6 &  | & 6\\
\end{array}
\right]
$$

row operations:

$$
\left[
\begin{array}{cccc}
1 & 2 & | & 2 \\
1 & 2 &  | & 2\\
\end{array}
\right]
$$

$$
\left[
\begin{array}{cccc}
1 & 2 & | & 2 \\
0 & 0 &  | & 0\\
\end{array}
\right]
$$

which gives the solution set

$x_1 = -2 \cdot t + 2$ and $x_2 = t$

which is expressed as a vectorial description of a straight line.

$$
\left[
\begin{array}{c}
x_1 \\
x_2
\end{array}
\right] =
\left[
\begin{array}{c}
2 \\
0
\end{array}
\right] + t \cdot
\left[
\begin{array}{c}
-2 \\
1
\end{array}
\right]
$$

**P.3.5**
Find the solution sets for the augemented matrices.

a)

$$
\left[
\begin{array}{cccc}
1 & -1 & -2 & | & 1 \\
-2 & 2 & 4 & | & -2 \\
3 & -3 & -6 & | & 3 
\end{array}
\right]
$$

row operations: R2/(-2) -> R2; R3/(3) -> R3

$$
\left[
\begin{array}{cccc}
1 & -1 & -2 & | & 1 \\
1 & -1 & -2 & | & 1 \\
1 & -1 & -2 & | & 1 
\end{array}
\right]
$$

row operations: R2 - R1 -> R2; R3 - R1 -> R3

$$
\left[
\begin{array}{cccc}
1 & -1 & -2 & | & 1 \\
0 & 0 & 0 & | & 0 \\
0 & 0 & 0 & | & 0 
\end{array}
\right]
$$

$x_2,\ x_3$ are free variables: $x_2 = t_2,\ x_3 = t_3$

$$
x_1 - t_2 -2 \cdot t_3 = 1
$$
